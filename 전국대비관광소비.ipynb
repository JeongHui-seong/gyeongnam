```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib
import matplotlib.ticker as ticker
import numpy as np
```

```python
fm.fontManager.addfont(path = "./font/NanumGothic.ttf")
matplotlib.rcParams["font.family"] = "NanumGothic"
```

```python
df = pd.read_csv("./csv/202310202503전국대비관광소비.csv")
```

```python
df
```

```python
x = np.arange(len(df['기준연월']))

fig, ax1 = plt.subplots(1, 1, figsize = (18, 8))

bar1 = ax1.bar(x - 0.2, df['전국 관광소비액(백만원)'], width = 0.4, label = '전국', color = 'lightgray')
ax1.set_ylabel('전국 관광소비액(백만원)', color = 'black')
ax1.tick_params(axis = 'y', labelcolor = 'black')
ax1.set_xticks(x)
ax1.set_xticklabels(df['기준연월'])
ax1.yaxis.set_major_formatter(ticker.ScalarFormatter(useOffset = False))
ax1.ticklabel_format(style = 'plain', axis = 'y')

ax2 = ax1.twinx()
bar2 = ax2.bar(x + 0.2, df['관광소비액(백만원)'], width = 0.4, label = '경상남도', color = 'green')
ax2.set_ylabel('경상남도 관광소비액(백만원)', color = 'green')
ax2.tick_params(axis = 'y', labelcolor = 'green')

fig.legend(loc = 'upper left', bbox_to_anchor = (0.85, 0.93))
plt.title('전국대비 관광소비액 비교')
plt.tight_layout()
plt.savefig("./img/전국대비관광소비액비교.png")
plt.show()
```
